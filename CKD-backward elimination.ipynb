```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
start =pd.read_csv('kidney_disease.csv')
start=start.iloc[:,1:]
start.drop('wc',axis=1,inplace=True)
start.drop('pcv',axis=1,inplace=True)
start.drop('rc',axis=1,inplace=True)
start
```

```python
start.isnull().any()
```

```python
start['age'].fillna((start['age'].mean()),inplace=True)
```

```python
start.isnull().any()
```

```python
start['bp'].fillna((start['bp'].mean()),inplace=True)
start['sg'].fillna((start['sg'].mean()),inplace=True)
start['al'].fillna((start['al'].mean()),inplace=True)
start['su'].fillna((start['su'].mean()),inplace=True)
start['bgr'].fillna((start['bgr'].mean()),inplace=True)
start['bu'].fillna((start['bu'].mean()),inplace=True)
start['sc'].fillna((start['sc'].mean()),inplace=True)
start['sod'].fillna((start['sod'].mean()),inplace=True)
start['pot'].fillna((start['pot'].mean()),inplace=True)
start['hemo'].fillna((start['hemo'].mean()),inplace=True)
```

```python
start['rbc']=start['rbc'].fillna(value=start['rbc'].mode().iloc[0])
start['dm']=start['dm'].fillna(value=start['dm'].mode().iloc[0])
start['htn']=start['htn'].fillna(value=start['htn'].mode().iloc[0])
start['ba']=start['ba'].fillna(value=start['ba'].mode().iloc[0])
start['pcc']=start['pcc'].fillna(value=start['pcc'].mode().iloc[0])
start['pc']=start['pc'].fillna(value=start['pc'].mode().iloc[0])
start['cad']=start['cad'].fillna(value=start['cad'].mode().iloc[0])
start['appet']=start['appet'].fillna(value=start['appet'].mode().iloc[0])
start['pe']=start['pe'].fillna(value=start['pe'].mode().iloc[0])
start['ane']=start['ane'].fillna(value=start['ane'].mode().iloc[0])
```

```python
start.isnull().any()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
start['rbc']=lb.fit_transform(start['rbc'])
start['rbc']
```

```python
start['pc']=lb.fit_transform(start['pc'])
start['pcc']=lb.fit_transform(start['pcc'])
start['ba']=lb.fit_transform(start['ba'])
start['htn']=lb.fit_transform(start['htn'])
start['dm']=lb.fit_transform(start['dm'])
start['cad']=lb.fit_transform(start['cad'])
start['appet']=lb.fit_transform(start['appet'])
start['pe']=lb.fit_transform(start['pe'])
start['ane']=lb.fit_transform(start['ane'])
start['classification']=lb.fit_transform(start['classification'])
```

```python
start
```

```python
x=start.iloc[:,0:21].values
y=start.iloc[:,21:].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import neighbors
```

```python
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
