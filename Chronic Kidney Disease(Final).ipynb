```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
start =pd.read_csv('kidney_disease.csv')
start=start.iloc[:,1:]
start.drop('age',axis=1,inplace=True)
start.drop('wc',axis=1,inplace=True)
start.drop('pcv',axis=1,inplace=True)
start.drop('rc',axis=1,inplace=True)
start.drop('su',axis=1,inplace=True)
start.drop('rbc',axis=1,inplace=True)
start.drop('pc',axis=1,inplace=True)
start.drop('sc',axis=1,inplace=True)
start.drop('pot',axis=1,inplace=True)
start.drop('dm',axis=1,inplace=True)
start.drop('cad',axis=1,inplace=True)
start.drop('pe',axis=1,inplace=True)
start
```

```python
start.isnull().any()
```

```python
start['bp'].fillna((start['bp'].mean()),inplace=True)
start['sg'].fillna((start['sg'].mean()),inplace=True)
start['al'].fillna((start['al'].mean()),inplace=True)
start['bgr'].fillna((start['bgr'].mean()),inplace=True)
start['bu'].fillna((start['bu'].mean()),inplace=True)
start['sod'].fillna((start['sod'].mean()),inplace=True)
start['hemo'].fillna((start['hemo'].mean()),inplace=True)
```

```python
start['htn']=start['htn'].fillna(value=start['htn'].mode().iloc[0])
start['ba']=start['ba'].fillna(value=start['ba'].mode().iloc[0])
start['pcc']=start['pcc'].fillna(value=start['pcc'].mode().iloc[0])
start['appet']=start['appet'].fillna(value=start['appet'].mode().iloc[0])
start['ane']=start['ane'].fillna(value=start['ane'].mode().iloc[0])
```

```python
start.isnull().any()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
start['pcc']=lb.fit_transform(start['pcc'])
start['ba']=lb.fit_transform(start['ba'])
start['htn']=lb.fit_transform(start['htn'])
start['appet']=lb.fit_transform(start['appet'])
start['ane']=lb.fit_transform(start['ane'])
start['classification']=lb.fit_transform(start['classification'])
```

```python
start
```

```python
x=start.iloc[:,0:12].values
y=start.iloc[:,12:].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
import pickle
pickle.dump(lr,open('decison2.pkl','wb'))
```

```python
y_pred=lr.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
